```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 데이터 설정
data1 = {
    'Year': [2020, 2021, 2022],
    'Size': [1468, 81625, 249648]
}

data2 = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Size': [34500, 101700, 188200, 300300, 376600]
}

data3 = {
    'Year': [2023, 2024, 2025, 2026, 2027],
    'Size': [448500, 507900, 561000, 610500, 660100]
}

# CAGR 계산 함수
def calculate_cagr(start_value, end_value, periods):
    return (end_value / start_value) ** (1 / periods) - 1

# 데이터프레임 생성
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)

# CAGR 계산
cagr1 = calculate_cagr(df1['Size'].iloc[0], df1['Size'].iloc[-1], len(df1['Year']) - 1)
cagr2 = calculate_cagr(df2['Size'].iloc[0], df2['Size'].iloc[-1], len(df2['Year']) - 1)
cagr3 = calculate_cagr(df3['Size'].iloc[0], df3['Size'].iloc[-1], len(df3['Year']) - 1)

# 시각화
sns.set(style="whitegrid")

def plot_data(df, cagr, title):
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Year', y='Size', data=df, palette=sns.dark_palette("#69d", reverse=True, as_cmap=True))
    plt.title(title, fontsize=20)
    plt.xlabel('Year')
    plt.ylabel('Market Size (in one million won)')
    plt.legend(labels=[f'CAGR: {cagr:.2%}'], loc='upper left')
    plt.show()

# 데이터1 그래프
plot_data(df1, cagr1, 'Market Size over Years')

# 데이터2 그래프
plot_data(df2, cagr2, 'Market Size over Years')

# 데이터3 그래프
plot_data(df3, cagr3, 'Market Size and Forecast')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 데이터 설정
data1 = {
    'Year': [2020, 2021, 2022],
    'Size': [14.68, 816.25, 2496.48]
}

data2 = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Size': [345, 1017, 1882, 3003, 3766]
}

data3 = {
    'Year': [2023, 2024, 2025, 2026, 2027],
    'Size': [4485, 5079, 5610, 6105, 6601]
}

# CAGR 계산 함수
def calculate_cagr(start_value, end_value, periods):
    return (end_value / start_value) ** (1 / periods) - 1

# 데이터프레임 생성
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)

# CAGR 계산
cagr1 = calculate_cagr(df1['Size'].iloc[0], df1['Size'].iloc[-1], len(df1['Year']) - 1)
cagr2 = calculate_cagr(df2['Size'].iloc[0], df2['Size'].iloc[-1], len(df2['Year']) - 1)
cagr3 = calculate_cagr(df3['Size'].iloc[0], df3['Size'].iloc[-1], len(df3['Year']) - 1)

# 시각화
sns.set(style="white")

def plot_data(df, cagr, title):
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x='Year', y='Size', hue='Year', dodge=False, data=df, palette='Blues', legend=False)
    for p in ax.patches:
        ax.annotate(format(p.get_height(), ',.0f'),
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha = 'center', va = 'center',
                    xytext = (0, 6),
                    textcoords = 'offset points')
    plt.title(title, fontsize=20)
    plt.xlabel('Year')
    plt.ylabel('Market Size (in one billion won)')
    plt.legend(labels=[f'CAGR: {cagr:.2%}'], loc='upper left')
    plt.show()

# 데이터1 그래프
plot_data(df1, cagr1, 'Meal Kits Market Size_Korea MFDS')

# 데이터2 그래프
plot_data(df2, cagr2, 'Meal Kits Market Size_Euromonitor')

# 데이터3 그래프
plot_data(df3, cagr3, 'Meal Kits Market Size and Forecast')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create the data dictionary
all_data = {
    'Year': [2020, 2021, 2022, 2023, 2024, 2025, 2026],
    'Market': [2778, 3102, 3123, 3247, 3439, 3630, 3845]
}

sns.set(style="white")

# Create the DataFrame
all_data = pd.DataFrame(all_data)

# Define colors for bars
palette = ['darkblue' if year <= 2023 else 'lightblue' for year in all_data['Year']]

# Calculate CAGR
start_value = all_data['Market'].iloc[0]
end_value = all_data['Market'].iloc[-1]
num_years = len(all_data['Year']) - 1
CAGR = ((end_value / start_value) ** (1 / num_years) - 1) * 100

# Create the plot
plt.figure(figsize=(12, 8))

# Plot the bar chart with Seaborn
bars = sns.barplot(x='Year', y='Market', data=all_data, palette=palette)

# Add the title and labels
plt.ylabel('Market Size (in billions of dollars)')
plt.xticks(range(len(all_data['Year'])), labels=['2020', '2021', '2022', '2023', '2024E', '2025E', '2026E'])  # Set X-axis labels to the exact years

# Add vertical dotted line after 2023
plt.axvline(x=3.5, color='grey', linestyle='--', linewidth=1)

# Set y-axis limits
plt.ylim(0, max(all_data['Market']) * 1.2)

# Add values on top of bars with increased font size
for index, row in all_data.iterrows():
    plt.text(index, row['Market'] + 50, round(row['Market'], 2), ha='center', va='bottom', fontsize=14)

# Remove the top and right spines (borders)
sns.despine(left=True, bottom=True)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a DataFrame from the provided data
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Market': [
        38790939006, 44448301187, 47327653596, 54028196292, 53736286078,
        56139024556, 59000502743, 61170907877, 63576595612, 65408225552,
        66119498181, 72454963683, 81833516908
    ]
}

df = pd.DataFrame(data)

# Converting the unit to trillion won (조 원)
df['Market'] = df['Market'] * 1000 / 1000000000000
df = df[df['Year'] >= 2012]

all_data = df

# Define colors for bars
palette = ['darkblue' if year <= 2022 else 'lightblue' for year in all_data['Year']]

# Calculate CAGR
start_value = all_data['Market'].iloc[0]
end_value = all_data['Market'].iloc[-1]
num_years = len(all_data['Year']) - 1
CAGR = ((end_value / start_value) ** (1 / num_years) - 1) * 100

# Create the plot
sns.set(style="white")
plt.figure(figsize=(12, 8))

# Plot the bar chart with Seaborn
bars = sns.barplot(x='Year', y='Market', data=all_data, palette=palette)

# Add the title and labels
plt.ylabel('Market Revenue (in trillions of won)')
plt.xticks(range(len(all_data['Year'])), labels=all_data['Year'], fontsize=15)  # Set X-axis labels to the exact years

# Set y-axis limits
plt.ylim(0, max(all_data['Market']) * 1.2)

# Add values inside bars with increased font size and bold font
for bar, label in zip(bars.patches, all_data['Market']):
    bar_height = bar.get_height()
    bars.annotate(format(label, '.1f'),  # Change to two decimal places
                  (bar.get_x() + bar.get_width() / 2, bar_height / 2),
                  ha='center', va='center',
                  fontsize=15, fontweight='bold', color='white')

# Remove the top and right spines (borders)
sns.despine(left=True, bottom=True)

plt.show()
```
